## <font color=#33A5FF>****Análisis de Datos de los Éxitos del K-pop 2020****<tfont>

El objetivo principal es identificar las tendencias generales que se produjeron en el K-Pop, específicamente en lo que respecta a ciertas características de audio.

<div class="well">
<strong>1.2 - </strong>Visualizar los Datos
</div>

```python
# AGRUPANDO POR ESTACION LAS COLUMNAS A USAR, Y SUMANDO LOS VALORES.

new_1 = kpop.groupby("artist/s")[["danceability","energy"]].sum()
new_2 = kpop.groupby("artist/s")[["speechiness","acousticness","valence"]].sum()
```

```python
# GRAFICA DE LINEA DE LOS VALORES LOCUASIDAD, ACÚSTICA Y VALENCIA POR ARTISTA

new_2.plot(figsize = (15,20), subplots = True)
plt.show()
```

```python
# Cambiar el tamaño del gráfico
fix, ax = plt.subplots(figsize = (15, 4))
new_1.plot(ax = ax)
plt.show()
```

```python
# Total de claves menores de 6 por cada artista
kpop_team = kpop[kpop['key'] > 6]
kpop_team['artist/s'].value_counts().plot(kind = 'bar',
                                          color = sns.color_palette('cubehelix'))
plt.show()                                             
```

```python
# Gráfico BoxPlot del índice de bailabilidad por artista
plt.figure(figsize = (13, 6))
sns.boxplot(x = 'artist/s', y = 'danceability', data = kpop)
plt.xticks(rotation = 90)
plt.show()
```

```python
#Top 10 artistas con mayor índice de energía
kpop_energy = kpop.sort_values(by = 'energy', ascending = False)[:10]
plt.figure(figsize = (20, 6))
plt.title('Top 10 artistas con mayor índice de energía', fontsize = 20)
c = sns.barplot(x = kpop_energy['artist/s'], y = kpop_energy['energy'],
             label = 'title', color = 'pink')
plt.ylabel('Índice de energía')
c.set_xticklabels(c.get_xticklabels(), rotation = 45)
c
plt.show()
      
```

```python
# Éxito que ha tenido cada artista con las canciones de 2020

nationality = kpop.groupby('artist/s').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize = (12, 6), color = sns.color_palette('magma'))
plt.show()
```

```python
# Grafica de Total del tiempo en sus canciones en segundos por cada artista
duration_by_artist = pd.DataFrame(kpop.groupby('artist/s', as_index = False)['duration_ms'].sum())
sns.set(style = 'whitegrid', color_codes = True)
ax = sns.barplot(x = 'artist/s', y = 'duration_ms', data = duration_by_artist.sort_values(by = 'duration_ms'), palette = 'Set2')
ax.set_xlabel('artist/s', fontsize = 30)
ax.set_ylabel('duration_ms', fontsize = 20)
plt.xticks(rotation = 75)

plt.rcParams['figure.figsize'] = (30, 10)
plt.title('Gráfica de Artista vs. Total de tiempo en sus canciones en segundos', fontsize = 20)
plt.show()
```

```python
# Grafica del total de sonoridad de cada artista
loudness_by_artist = pd.DataFrame(kpop.groupby('artist/s', as_index = False)['loudness'].sum())
sns.set(style = 'whitegrid', color_codes = True)
ax = sns.barplot(x = 'artist/s', y = 'loudness', data = loudness_by_artist.sort_values(by = 'loudness'), palette = 'rocket')
ax.set_xlabel('Artist/s', fontsize = 30)
ax.set_ylabel('loudness', fontsize = 20)
plt.xticks(rotation = 75)

plt.rcParams['figure.figsize'] = (20, 10)
plt.title('Gráfica de Artista vs. Total de Sonoridad', fontsize = 20)
plt.show()
```
